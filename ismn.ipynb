```python
from ismn.interface import ISMN_Interface
import pandas as pd

# Initialize ISMN interface
ismn_data = ISMN_Interface("E:/UCDavis/QE/ISMN_archive_20240205", parallel=False)

# Define the geographical bounds of California (approximate)
min_lat, max_lat = 32.5, 42.0
min_lon, max_lon = -124.5, -114.1
max_depth = 0.1  # Maximum depth

# Output CSV file path
output_csv_path = 'E:/UCDavis/QE/ismn/california_scan_soil_moisture.csv'

# Aggregate SCAN soil moisture data for stations in California
all_data = []
```

```python
# Access SCAN network
scan_network = ismn_data['SCAN']

print(scan_network)
```

```python
sensor = ismn_data['SCAN']['AshValley']['Hydraprobe-Digital-Sdi-12-(2.5-Volt)_soil_moisture_0.050800_0.050800']
pd = sensor.metadata.to_pd()
```

```python
print(pd)
```

```python
print(pd[('timerange_from')])
print(pd[('timerange_to')])
```

```python
# Iterate over stations in the SCAN network
for station_name in scan_network.stations:
    station = scan_network[station_name]
    station_meta = station.metadata.to_pd()

    # Access latitude and longitude from metadata
    latitude = station_meta.loc[('latitude', 'val')].item()
    longitude = station_meta.loc[('longitude', 'val')].item()

    # Check if the station is within California
    if not np.isnan(latitude) and not np.isnan(longitude):
        if min_lat <= latitude <= max_lat and min_lon <= longitude <= max_lon:
            for sensor_name in station.sensors:
                sensor = station[sensor_name]
                if 'soil_moisture' in sensor.variable:
                    sensor_meta = sensor.metadata.to_pd()

                    # Extract depth information
                    depth_from = sensor_meta[('variable', 'depth_from')]
                    depth_to = sensor_meta[('variable', 'depth_to')]

                    if depth_from <= max_depth and depth_to <= max_depth:
                        try:
                            data = sensor.data  # Use the data directly
                            # Filter out rows where 'soil_moisture_flag' is not 'G'
                            data = data[data['soil_moisture_flag'] == 'G']
                            
                            if not data.empty:
                                data['station'] = station_name
                                data['network'] = 'SCAN'
                                data['depth_from'] = depth_from
                                data['depth_to'] = depth_to
                                data['latitude'] = latitude
                                data['longitude'] = longitude
                                all_data.append(data)
                        except Exception as e:
                            print(f"Error reading data for sensor {sensor.name} at station {station_name}: {e}")
```

```python
print(all_data)
```

```python
from datetime import datetime

# Convert string to datetime object
start_date = datetime(2015, 6, 27)

# Filter each DataFrame in all_data for dates after 2015-06-27 and add a datetime column
filtered_data = []
for df in all_data:
    df = df[df.index > start_date]  # Apply date filter
    if not df.empty:
        df['datetime'] = df.index  # Add datetime column
        filtered_data.append(df)

print(filtered_data)
```

```python
# Function to split data into multiple files
def export_to_csv(filtered_data, base_output_path, max_rows_per_file=500000):
    file_count = 1
    for i in range(0, len(filtered_data), max_rows_per_file):
        output_path = f"{base_output_path}_part{file_count}.csv"
        chunk = filtered_data[i:i + max_rows_per_file]
        chunk.to_csv(output_path, index=False)
        print(f"Data exported to {output_path}")
        file_count += 1

# Check if filtered_data is not empty before processing
if filtered_data:
    combined_data = pd.concat(filtered_data, ignore_index=True)
    export_to_csv(combined_data, output_csv_path, max_rows_per_file=500000)
else:
    print("No data meets the criteria.")
```
